```python
#1 Contagem Crescente
# Usando for e range(), exiba os números de 1 até 10, um por linha.

for i in range(1, 6):
    print(i)
for i in range(6, 11):
    print(i)
```

```python
#2 Contagem Regressiva
# Usando for e range(), exiba os números de 10 até 1 em ordem decrescente, um por linha.

for i in range(10, 0, -1):
    print(i)
```

```python
#3 Tabuada do Cogumelo
# Leia um número inteiro.
#Usando for e range(), exiba a tabuada completa desse número, do 1 ao 10, no formato: 'N x I = R'.

num = int(input("Digite um numero"))
for i in range(11):
    resultado = num * i
    print(f"{num} x {i} = {resultado}")
```

```python
#4 Soma da Sequência
#Leia um número inteiro N positivo. Usando for, calcule e exiba a soma de todos os números de 1 até N.
#Valide que N é positivo

num = int(input("Digite um numero: "))
if num > 0:
    soma = 0
    for i in range(1, num+1):
        soma+= i
    print(soma)

else:
    print("Numero negativo")
```
