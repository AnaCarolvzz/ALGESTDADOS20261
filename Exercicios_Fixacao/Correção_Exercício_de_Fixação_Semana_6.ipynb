```python
#4 - Leia um número inteiro N positivo.
#Usando for, calcule e exiba a soma de todos os números de 1 até N.
#Valide que N é positivo.

num = int(input('Digite um número: ')) #Leia um número inteiro N positivo.
if num > 0: #Valide que N é positivo.
    #calcule e exiba a soma de todos os números de 1 até N.
    soma = 0
    for i in range(1,num+1): #for i in range (1,5): # 1 a 4
        soma+= i #soma = soma + i
    print(soma)
else:
    print('Número negativo')

```

```python
#5 - Leia um número inteiro N positivo.
#Usando for e range(),
#exiba todos os números pares de 0 até N.
#Ao final, exiba a quantidade total de pares encontrados.

n = int(input('Digite um número positivo:'))
contador = 0
if n > 0:
    for i in range(0,n+1,2):
        print(i)
        contador+=1
    print(f'Tem {contador} pares')
else:
    print('Número não positivo')
```

```python
#6 - Leia um número N de alunos (positivo).
#Usando for, leia a nota de cada aluno, valide que está entre 0 e 10,
#acumule a soma e calcule a média ao final.
#Exiba a média com 2 casas decimais.

#Leia um número N de alunos (positivo).
n = int(input('Digite a quantidade de alunos:'))

#Valida que a quantidade é positiva
if n > 0:
    soma = 0 #inicializa a variável soma
    for i in range(1,n+1): #percorre do 1° ao último aluno
        #Adiciona a nota de cada aluno
        nota = float(input(f'Digite a nota do {i}° aluno:'))

        #Se a nota for negativa ou acima de 10, é considerado 0 (zero)
        if 0 <= nota <= 10:
            #se a nota for válida é somada
            soma+=nota
        else:
            #senão, conta apenas o aluno e a nota fica zerada
            print('A nota foi considerada como 0 (zero)')

    media = soma/n #calcula a média

    print(f'A média de notas dos {n} alunos é: {media:.2f}') #imprime a média final

#Caso o valor de alunos seja negativo exibe o erro
else:
    print('Número não positivo')
```

```python
#08 - Maior e Menor
#Leia um número N de valores (N >= 2).
#Usando for, leia cada valor e, ao final, exiba qual foi o maior e o menor valor digitado.
#Use variáveis descritivas como maior_valor e menor_valor.

#Leia um número N de valores (N >= 2).
N_valores = int(input('Digite um valor maior ou igual a 2: '))
#Verifica se o N_valores é maior ou igual a 2.
if(N_valores >=2):

    #Repetir até chegar em N_valores
    for i in range(N_valores):

        #Para cada iteração vai pedir pra digitar um novo valor inteiro
        valor = int(input(f'Digite o {i+1}° valor: '))

        #Se for a primeira leitura, ela inicializa as variáveis menor e maior
        if i == 0:
            maior = valor
            menor = valor

        #Nas demais, verifica se tem um novo menor
        if valor <= menor:
            menor = valor

        #Ou também verifica se tem um novo maior
        if valor >= maior:
            maior = valor

        print(f'Na {i+1}° iteração o Maior: {maior} / Menor: {menor}') #Apenas para entendimento (debug)

    #Termina de ler e comparar todos os números e imprime
    print(f'Maior: {maior} / Menor: {menor}')

else: #se, não
    #Imprime o Erro
    print('Valor inválido')
```

```python
#09 Estrelas em Escada
#Leia um número inteiro N (1 a 10).
#Usando for com range() aninhado, exiba uma escada de asteriscos:
#a linha 1 tem 1 estrela, a linha 2 tem 2, até a linha N.

asteriscos =  int(input('Digite um valor de 1 a 10: '))
if 1<= asteriscos <=10:
    palavra = ''
    for i in range(asteriscos):
        palavra = palavra + '*'
        print(palavra)
else:
    print('ERRRRRROU!')
```

```python
#11
#Leia um número N positivo.
#Usando for, leia N números inteiros.
#Para cada número, classifique-o como positivo, negativo ou zero e exiba a classificação.
#Ao final, exiba a contagem de cada tipo.

#parte 1
numero = int(input('Digite um número: '))
if numero >= 0:
    #parte 2
    conta_positivo = 0
    conta_negativo = 0
    conta_zero = 0
    for numeros in range(numero):
        inteiros = int(input(f'Digite o {numeros+1}° número: '))
        #parte 3
        if inteiros > 0:
            conta_positivo+=1
            print(f'{inteiros} - positivo')
        elif inteiros < 0:
            conta_negativo+=1
            print(f'{inteiros} - negativo')
        else:
            conta_zero+=1
            print('Valor é Zero')
#parte 4
    print(f'Positivos: {conta_positivo}')
    print(f'Negativos: {conta_negativo}')
    print(f'Zeros: {conta_zero}')
```

```python
#13
#Leia um número inteiro N > 1.
#Usando for e range(), verifique se N é primo (divisível apenas por 1 e por ele mesmo).
#Use uma variável booleana eh_primo.
#Exiba o resultado.

#Parte 1
numero = int(input('Digite um número maior que 1:'))
if numero > 1:
    #Parte 2
    conta_divisor = 0
    eh_primo = False #Parte 3 - Declara tudo como falso até que prove o contrário
    for divisor in range(1,numero+1):
        if numero%divisor == 0:
            conta_divisor+=1
    if conta_divisor == 2: #Só pode ser dividido por 1 ou ele mesmo
        eh_primo = True #Valida como Primo

print(f'Número é primo? {eh_primo}') #Parte 4
```

```python
#15
#Leia o nome e N notas (N definido pelo usuário, mínimo 2) de um aluno.
#Usando for, leia e valide cada nota (0-10). Calcule a média.
#Com if/elif, determine o conceito (A >= 9, B >= 7, C >= 5, D >= 3, E abaixo).
#Exiba nome, média com 2 casas e conceito.

#Parte 1
nome = input('Digite seu nome: ')
quant_notas = int(input('Quantidade de notas (mín. 2): '))
if quant_notas >= 2:
#Parte 2
    soma = 0
    for quant in range(quant_notas): #Captura todas as notas
        nota = float(input(f'Digite a {quant+1}ª nota:'))
        if 0<= nota <=10:
            soma+=nota #Soma nota válida.
        else:
            soma+=0 #Nota inválida é somada como zero.

    media = soma/quant_notas #ao fim, calcula a média final

#Parte 3
    if media >= 9:# 10 >= nota >= 9
        print(f'{nome} Conceito A - {media:.2f} pontos')
    elif media >= 7:# 9 > nota >= 7
        print(f'{nome} Conceito B - {media:.2f} pontos')
    elif media >= 5:# 7 > nota >= 5
        print(f'{nome} Conceito C - {media:.2f} pontos')
    elif media >= 3:# 5 > nota >= 3
        print(f'{nome} Conceito D - {media:.2f} pontos')
    else:# 3 > nota >= 0
        print(f'{nome} Conceito E - {media:.2f} pontos')

else: #Caso não tenha 2 ou mais notas
    print('Notas mínimas não alcançadas!')
```
