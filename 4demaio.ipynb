```python
turma = ['joao', 'arthur', 'carol', 'isa']
#a
print('letra A:')
for aluno in turma:
    print(f'presente: {aluno}')
turma.append('ana')
#b
print('\nletra b:')
for aluno in turma:
    print(f'presente: {aluno}')
```

```python
valores = [8, 7, 6, 5, 4, 3]
print(valores)
valores.remove(6)
print(valores)
valores.pop()
print(valores)
valores.sort()
print(valores)
for i, valores in enumerate(valores):
  print(f'i valores: {valores}')
```
